```python
# path = os.getcwd()
kumulativna = 0

for (p, dirs, files) in os.walk(path):
    print(p)

    for i in dirs:
        print(f'[{i}]')

    for i in files:
        print(i, end=" ")
        konkretny = os.path.join(p,i)
        velkost = os.path.getsize(konkretny)
        print(velkost)
        kumulativna = kumulativna + velkost
        
    print()
    print("velkost adresára:", kumulativna)
```
